### Lab session – S-metaheuristic algorithm - the Tabu Search algorithm (TS)
### TSP problem

#### - Version 1, only Tabu list 

```python
# load data from tsp file
import numpy as np

tsp_data = np.loadtxt('../TSPDataset/gr17.2085.tsp')
print(tsp_data.shape)
```

```python
import math
import queue

def initial_solution(length):
    s = np.random.choice(range(length), length, replace=False) # cities in [0,length]
    return s

def objectivefunction(s):
    cost = 0
    for i in range(s.shape[0]-1):
        cost = cost + tsp_data[s[i]][s[i+1]]
    cost = cost + tsp_data[s[-1]][s[0]]
    return cost

def pick_next_nontabu_neighbor(neighbors_list, tabu_list):
    # on cherche le premier voisin qui n'est pas dans la liste tabu
    for neighbor in neighbors_list:
        if not any((neighbor == tabu).all() for tabu in list(tabu_list.queue)):
            return neighbor
    # None si aucun voisin non-tabu n'est trouvé
    return None

def tabu_neighbor(s, tabu_list):
    # on veérifie si la solution s est présente dans la liste tabu
    
    for tabu in list(tabu_list.queue): # on parcourt tous les éléments de la liste tabu
        if (s == tabu).all(): # true si tous les éléments de s sont égaux à ceux de tabu
            return True
    return False

def generate_sorted_neighbors(s):
    # on génere tous les voisins de la solution actuelle en échangeant deux villes
    
    # on initilialise les listes des voisins et leurs coûts
    neighbors = []
    costs = []
    
    for i in range(len(s)): # on parcourt toutes les villes
        for j in range(i+1, len(s)): # on parcourt les villes suivantes pour échanger
            
            # ON crée un nouveau voisin en échangeant les villes aux indices i et j
            neighbor = np.copy(s)
            
            neighbor[i], neighbor[j] = neighbor[j], neighbor[i] # echanger les villes
            
            # on ajoute le voisin à la liste des voisins
            neighbors.append(neighbor)
            
            # calcul du coût du voisin
            costs.append(objectivefunction(neighbor))
            
    # on trie les voisins par leurs coûts en ordre croissant
    sorted_indices = np.argsort(costs)
    
    # on retourne les voisins et leurs coûts triés
    neighbors = [neighbors[i] for i in sorted_indices]
    costs = [costs[i] for i in sorted_indices]
    return neighbors, costs

def best_admessible_neighbor(s, tabu_list):
    # return sorted neighbors_list and associated costs
    neighbors_list, costs = generate_sorted_neighbors(np.copy(s))
    
    # les indices 0 contiennent le meilleur voisin
    best_neighbor, cost = neighbors_list[0], costs[0]
    
    # check if the best neighbor is a tabu
    tabu = tabu_neighbor(best_neighbor, tabu_list)
    
    if (cost < objectivefunction(s)):
        return best_neighbor, tabu
    else: # escape from local optimum by accepting a degraded solution
        if not tabu:
            return best_neighbor, tabu
        else: # pick non-tabu neighbor
            return pick_next_nontabu_neighbor(neighbors_list, tabu_list), False
```

### Best solution and associated trajectory curve for each version

```python
import queue
import matplotlib.pyplot as plt

tabu_list = queue.Queue(5) # FIFO queue of 5 items max

costs_hist = []

# Random initial solution and its cost
length_s = 17
s = initial_solution(length_s) #np.random.choice(range(length_s), length_s, replace=False) #cities in [0,length_s]
cost_s = objectivefunction(s)

# Update tabu list
tabu_list.put(np.copy(s))

# Save history of costs
costs_hist.append(cost_s)

# Consider solution s as the best solution for now
best_s = np.copy(s) # hard copy
best_cost = cost_s

total_it = 0 # max iterations

while total_it < 100:
    adm_s, tabu = best_admessible_neighbor(np.copy(s), tabu_list)
    cost_adm_s = objectivefunction(adm_s)
    costs_hist.append(cost_adm_s)

    # update s
    s = np.copy(adm_s)

    if not tabu: # adm_s not satisfying an aspiration criterion
        # update best solution if necessary
        if cost_adm_s < best_cost: 
            best_s = np.copy(adm_s)
            best_cost = cost_adm_s

        # update tabu list by adding adm_s as a visited solution
        
        if not tabu_list.full():
            # on vérifie si la liste tabu n'est pas pleine car on ne peut pas ajouter plus de 5 éléments
            tabu_list.put(np.copy(adm_s))
        else:
            
            # si elle est pleine :on retire le premier élément de la liste tabu et on ajoute le nouvel élément
            tabu_list.get()
            tabu_list.put(np.copy(adm_s))

    total_it += 1 # increment iteration

print('best sol, cost {} {}'.format(best_s, best_cost))
```

```python
import matplotlib.pyplot as plt

plt.plot(costs_hist, 'b')
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.show()
```

### - Version 2, add intensification process

```python
import queue

import matplotlib.pyplot as plt

def intensify_search(best_s, tabu_list, intensification_factor=5):
    for _ in range(intensification_factor):
        neighbors, _ = generate_sorted_neighbors(best_s)
        for neighbor in neighbors:
            if not tabu_neighbor(neighbor, tabu_list):
                return neighbor
    return best_s

tabu_list = queue.Queue(5)  # FIFO queue of 5 items max
costs_hist = []

# Random initial solution and its cost
length_s = 17
s = initial_solution(length_s)
cost_s = objectivefunction(s)

# Update tabu list
tabu_list.put(np.copy(s))

# Save history of costs
costs_hist.append(cost_s)

# Consider solution s as the best solution for now
best_s = np.copy(s)  
best_cost = cost_s

total_it = 0  # max iterations
intensification_interval = 20  # interval for intensification

while total_it < 100:
    if total_it % intensification_interval == 0:
        s = intensify_search(best_s, tabu_list)
    
    adm_s, tabu = best_admessible_neighbor(np.copy(s), tabu_list)
    cost_adm_s = objectivefunction(adm_s)
    costs_hist.append(cost_adm_s)

    # update s
    s = np.copy(adm_s)

    if not tabu:  # adm_s not satisfying an aspiration criterion
        # update best solution if necessary
        if cost_adm_s < best_cost:
            best_s = np.copy(adm_s)
            best_cost = cost_adm_s

        # update tabu list by adding adm_s as a visited solution
        if not tabu_list.full():
            tabu_list.put(np.copy(adm_s))
        else:
            tabu_list.get()
            tabu_list.put(np.copy(adm_s))

    total_it += 1  # increment iteration

print('best sol, cost {} {}'.format(best_s, best_cost))

# Plot the cost history
plt.plot(costs_hist, 'b')
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.show()

# Plot the best solution over cities
fig, ax = plt.subplots(1, 1, figsize=(4, 4))
ax.plot(mds_coords[best_s, 0], mds_coords[best_s, 1], 'b-o')
plt.plot([mds_coords[best_s[0], 0], mds_coords[best_s[-1], 0]],
         [mds_coords[best_s[0], 1], mds_coords[best_s[-1], 1]], 'b-o')
plt.show()
```

### - Version 3, add diversification process

```python
def diversify_search(tabu_list, diversification_factor=5):
    # Generate a new random solution that is not in the tabu list
    for _ in range(diversification_factor):
        new_solution = initial_solution(length_s)
        if not tabu_neighbor(new_solution, tabu_list):
            return new_solution
    return initial_solution(length_s)

tabu_list = queue.Queue(5)  
costs_hist = []

# Random initial solution and its cost
length_s = 17
s = initial_solution(length_s)
cost_s = objectivefunction(s)

# Update tabu list
tabu_list.put(np.copy(s))

# Save history of costs
costs_hist.append(cost_s)

best_s = np.copy(s)  
best_cost = cost_s

total_it = 0  # max iterations
diversification_interval = 20  # interval for diversification

while total_it < 100:
    if total_it % diversification_interval == 0:
        s = diversify_search(tabu_list)
    
    adm_s, tabu = best_admessible_neighbor(np.copy(s), tabu_list)
    cost_adm_s = objectivefunction(adm_s)
    costs_hist.append(cost_adm_s)

    # update s
    s = np.copy(adm_s)

    if not tabu:  # adm_s not satisfying an aspiration criterion
        # update best solution if necessary
        if cost_adm_s < best_cost:
            best_s = np.copy(adm_s)
            best_cost = cost_adm_s

        # update tabu list by adding adm_s as a visited solution
        if not tabu_list.full():
            tabu_list.put(np.copy(adm_s))
        else:
            tabu_list.get()
            tabu_list.put(np.copy(adm_s))

    total_it += 1 

print('best sol, cost {} {}'.format(best_s, best_cost))

# Plot the cost history
plt.plot(costs_hist, 'b')
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.show()

# Plot the best solution over cities
fig, ax = plt.subplots(1, 1, figsize=(4, 4))
ax.plot(mds_coords[best_s, 0], mds_coords[best_s, 1], 'b-o')
plt.plot([mds_coords[best_s[0], 0], mds_coords[best_s[-1], 0]],
         [mds_coords[best_s[0], 1], mds_coords[best_s[-1], 1]], 'b-o')
plt.show()
```

## Conclusion :

L'algorithme de recherche Tabu est une méthode d’optimisation avancée qui permet d’éviter de revisiter les mêmes solutions en mémorisant un historique des dernières explorations. Ici,, on a implémenté 3 versions pour le problème TSP.

Etapes de l'algorithme de recherche Tabu :
1. Explorer les voisins d'une solution actuelle.
2. Éviter les cycles en mémorisant les solutions déjà visitées dans une liste Tabu.
3. Améliorer la recherche en ajoutant des stratégies d’intensification et de diversification pour obtenir de meilleures solutions plus rapidement.


### Version 1 : Implémentation de base avec la liste Tabu

#### ⚙️ Fonctionnement
1. Génération d’une solution initiale avec `initial_solution(length)`.
2. Calcul du coût de la solution avec `objectivefunction(s)`.
3. Génération des voisins triés selon leur coût via `generate_sorted_neighbors(s)`.
4. Choix du meilleur voisin admissible via `best_admessible_neighbor(s, tabu_list)`.
5. Mise à jour de la liste Tabu :
    - Si un voisin est Tabu, on tente d’en choisir un autre via `pick_next_nontabu_neighbor(neighbors_list, tabu_list)`.
    - Sinon, on l'ajoute à la liste avec `tabu_list.put(np.copy(adm_s))`.
6. Répétition du processus jusqu'à atteindre un nombre maximal d'itérations.

#### 📌 Points clés
- ✅ Empêche de revisiter les solutions récentes.
- ✅ Optimisation progressive vers une meilleure solution.
- ⚠️ Peut rester bloqué dans un optimum local si l’on ne diversifie pas la recherche.

### Version 2 : Ajout de l’Intensification 

L’intensification permet de se focaliser sur les meilleures solutions trouvées afin d’explorer plus en profondeur des zones prometteuses de l’espace de recherche.

#### ⚙️ Fonctionnement
1. Toutes les 20 itérations, on applique la fonction `intensify_search(best_s, tabu_list)`.
2. Cette fonction explore plus en détail les voisins non Tabu pour améliorer la solution actuelle.
3. Si une meilleure solution est trouvée, elle est adoptée.
4. On reprend ensuite le processus classique de la liste Tabu.

#### 📌 Points clés
- ✅ Permet d'exploiter pleinement les bonnes solutions.
- ✅ Augmente les chances d'obtenir une solution optimale.
- ⚠️ Peut toujours être piégé dans un optimum local si la recherche est trop focalisée.

### Version 3 : Ajout de la Diversification 

Si l’algorithme reste bloqué dans un optimum local, il faut parfois explorer de nouvelles régions pour éviter de tourner en rond. C’est le rôle de la diversification.

#### ⚙️ Fonctionnement
1. Toutes les 20 itérations, on applique `diversify_search(tabu_list)`.
2. Cette fonction génère une nouvelle solution aléatoire qui n’est pas dans la liste Tabu.
3. L’algorithme repart alors dans une direction totalement différente.
4. On poursuit ensuite l’exploration avec la liste Tabu classique.

#### 📌 Points clés
- ✅ Évite de stagner dans une zone de l’espace de recherche.
- ✅ Permet une meilleure couverture des solutions possibles.
- ⚠️ Peut ralentir l’optimisation si appliqué trop souvent.
